```python
import numpy as np
import matplotlib.pyplot as plt
```

Matplotlib Data Visualisation

```python
x = np.linspace(0,5,11)

y = x ** 2

print(x)
print(y)
```

```python
# Functional method and object oriented method

# functional
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
# plt.subplot(rows,column,index)

plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

plt.show()
```

```python
# object oriented

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'r')

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')

plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes1.set_title('Larger plot')

axes2.plot(y,x,'b')
axes2.set_title('Smaller plot')

plt.show()
```

```python
# subploting

fig,axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(y,x)


plt.tight_layout()
```

```python
#axes is a list type object
axes
```

Figure Size and DPI

```python
fig = plt.figure(figsize=(3,4))

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize=(3,4))

axes.plot(x,y)
plt.show()
```

```python
fig,axes = plt.subplots(2,1,figsize=(3,4),dpi=100)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
plt.show()
```

```python
fig.savefig('my_fig.png')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,x**2,label='X squared')
axes.plot(x,x**3,label='X cubed')

axes.legend(loc=0)
```
