```python
import numpy as np
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights=sns.load_dataset('flights')
flights.head()
```

```python
tips.head()
```

Matrix Plot

```python
tc = tips.corr()
tc
```

```python
sns.heatmap(tc,annot=True)
```

```python
ptdf = flights.pivot_table(index='month',columns='year',values='passengers')
ptdf
```

```python
sns.heatmap(ptdf)
```

```python
sns.heatmap(ptdf,cmap='magma',linecolor='white',linewidth='1')
```

```python
sns.clustermap(ptdf,cmap='coolwarm',linecolor='green',linewidth=0.5,standard_scale=1)
```
